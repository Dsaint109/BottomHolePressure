```python
import csv
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# For compatibility across multiple platforms
import os
IB = os.environ.get('INSTABASE_URI',None) is not None
open = ib.open if IB else open
```

```python
wells = []
with open('results.csv', 'r') as F:
    reader = csv.DictReader(F)
    for row in reader:
        wells.append(row)
```

```python
for well in wells:
    plt.plot([0, float(well["H"])], [float(well["Pwh"]), float(well["P_ideal"])], 'b-', label="Ideal case")
    plt.plot([0, float(well["H"])], [float(well["Pwh"]), float(well["P_real"])], 'r-', label="Real case")
    plt.plot([0, float(well["H"])], [float(well["Pwh"]), float(well["Pbh"])], 'g-', label="Bottom Hole pressure")
    plt.ylabel('Pressure')
    plt.xlabel('Depth')
    plt.show()
```
